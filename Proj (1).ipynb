```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
import numpy as np

# Define your column groups
continuous_cols = ['popularity', 'runtime', 'rate']  # 3 continuous columns
categorical_cols = ['date']  # 1 discrete/categorical column (will be one-hot encoded)

# Create the preprocessor
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), continuous_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), categorical_cols)
    ],
    remainder='passthrough',
    verbose_feature_names_out=False  # Disable prefixes!
  # passthrough any other columns
)
# Transform data
data_transformed = preprocessor.fit_transform(features)

# Get feature names after transformation
feature_names = preprocessor.get_feature_names_out()

# Convert back to DataFrame
data_df = pd.DataFrame(data_transformed, columns=feature_names, index=data.index)

# Display
data_df
```

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# Define a function to build an autoencoder
def build_autoencoder(input_dim, encoding_dim):
    """Builds a simple autoencoder model."""
    input_layer = tf.keras.layers.Input(shape=(input_dim,), name='Input_Layer')

    # The encoder layer
    x = tf.keras.layers.Dense(128, activation='relu')(input_layer)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Dropout(0.2)(x)
    x = tf.keras.layers.Dense(64, activation='relu')(x)
    encoded = tf.keras.layers.Dense(encoding_dim, activation='relu', name='Encoder_Layer')(x)
    # The decoder layer
    x = tf.keras.layers.Dense(64, activation='relu')(encoded)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Dropout(0.2)(x)
    x = tf.keras.layers.Dense(128, activation='relu')(x)
    decoded = tf.keras.layers.Dense(input_dim, activation='sigmoid', name='Decoder_Layer')(x)
    
    autoencoder = tf.keras.models.Model(inputs=input_layer, outputs=decoded, name='Autoencoder')
    autoencoder.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss='mse')
    return autoencoder

# Experiment with different encoding dimensions
encoding_dims = [8, 16, 32, 64, 128]
losses = []

for dim in encoding_dims:
    autoencoder = build_autoencoder(input_dim=data_transformed.shape[1], encoding_dim=dim)
    # Convert features to TensorFlow tensors with explicit dtype
    features_tensor = tf.convert_to_tensor(data_transformed, dtype=tf.float32)
    history = autoencoder.fit(features_tensor, features_tensor, epochs=20, batch_size=32, verbose=0, validation_split=0.2)
    val_loss = history.history['val_loss'][-1]
    losses.append(val_loss)
    print(f"encoding_dim={dim}, validation_loss={val_loss:.4f}")

# Plot the results
plt.plot(encoding_dims, losses, marker='o')
plt.xlabel('Encoding Dimension')
plt.ylabel('Validation Loss')
plt.title('Validation Loss vs Encoding Dimension')
plt.grid()
plt.show()
```

```python
import tensorflow as tf


# Define the autoencoder
input_dim = data_transformed.shape[1]  # Number of features
encoding_dim = 32  # Size of the compressed representation

def build_autoencoder(input_dim, encoding_dim):
    """Builds a simple autoencoder model."""
    input_layer = tf.keras.layers.Input(shape=(input_dim,), name='Input_Layer')

    # The encoder layer
    x = tf.keras.layers.Dense(128, activation='relu')(input_layer)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Dropout(0.2)(x)
    x = tf.keras.layers.Dense(64, activation='relu')(x)
    encoded = tf.keras.layers.Dense(encoding_dim, activation='relu', name='Encoder_Layer')(x)
    # The decoder layer
    x = tf.keras.layers.Dense(64, activation='relu')(encoded)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Dropout(0.2)(x)
    x = tf.keras.layers.Dense(128, activation='relu')(x)
    decoded = tf.keras.layers.Dense(input_dim, activation='sigmoid', name='Decoder_Layer')(x)
    
    autoencoder = tf.keras.models.Model(inputs=input_layer, outputs=decoded, name='Autoencoder')
    autoencoder.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss='mse')
    return autoencoder

print(f"--- Training final autoencoder with encoding_dim={encoding_dim} ---")
final_autoencoder = build_autoencoder(input_dim=input_dim, encoding_dim=encoding_dim)
early_stop = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss', patience=10, restore_best_weights=True
)
final_autoencoder.fit(data_transformed,data_transformed
                      ,epochs=100,
                      batch_size=32,
                      shuffle=True,
                      validation_split=0.2,
                      callbacks=[early_stop]
)
# Encoder model to extract feature embeddings
encoder = tf.keras.models.Model(
    inputs=final_autoencoder.input,
    outputs=final_autoencoder.get_layer('Encoder_Layer').output,
    name='Encoder')
# Save the autoencoder model
# autoencoder.save("autoencoder_model.h5")
# Save the encoder model
encoder.save("encoder_model_last.keras", save_format="keras")
```

```python
# === ADD THIS CODE TO THE END OF YOUR NOTEBOOK ===
import Dummies
import joblib
import tensorflow as tf
import Scrap2 
import numpy as np
import pandas as pd
preprocessor = joblib.load('preprocessor.pkl')
from sentence_transformers import SentenceTransformer
SBERT_model = SentenceTransformer('all-MiniLM-L6-v2')
encoder = tf.keras.models.load_model("encoder_model_last.keras")

# We are re-creating the exact steps for a single input
d = Scrap2.fetch_movie_data_for_processing('until dawn')
original_title, genres, lang, overview, popularity, date_str, runtime, rate = d[0], d[1], d[2], d[3], d[4], d[5], d[6], d[7]

POPULARITY_LIMIT = 12.0
RUNTIME_MIN = 60.0
RUNTIME_MAX = 200.0

# 1. Clip Popularity
if popularity > POPULARITY_LIMIT:
    print(f"Clipping popularity for '{original_title}' from {popularity} to {POPULARITY_LIMIT}")
    popularity = POPULARITY_LIMIT

# 2. Clip Runtime
if runtime < RUNTIME_MIN:
    print(f"Clipping runtime for '{original_title}' from {runtime} to {RUNTIME_MIN}")
    runtime = RUNTIME_MIN
elif runtime > RUNTIME_MAX:
    print(f"Clipping runtime for '{original_title}' from {runtime} to {RUNTIME_MAX}")
    runtime = RUNTIME_MAX

    
def dateSTR(x): return 'Old' if x < 2000 else 'Modern' if 2000 < x < 2010 else 'New'
date_val = int(date_str.split('-')[0])
date_enc = np.array(dateSTR(date_val)).reshape(1, -1)

lang_enc = Dummies.selectdummyLanguage(lang).reshape(1, -1)
genres_enc = Dummies.selectdummyGenre(genres).reshape(1, -1)

popularity = np.array(popularity).reshape(1, -1).astype('float32')
runtime = np.array(runtime).reshape(1, -1).astype('float32')
rate = np.array(rate).reshape(1, -1).astype('float32')

arrvalues = np.hstack([popularity, runtime, rate, date_enc, lang_enc, genres_enc])



language_cols = Dummies.LANGUAGE_COLUMNS_LIST
genre_cols = Dummies.GENRE_COLUMNS_LIST

cols = ['popularity', 'runtime', 'rate', 'date'] + language_cols + genre_cols

prepared_data = pd.DataFrame(arrvalues, columns=cols)
preprocessed_data = preprocessor.transform(prepared_data).astype('float32')
preprocessed_data


```
