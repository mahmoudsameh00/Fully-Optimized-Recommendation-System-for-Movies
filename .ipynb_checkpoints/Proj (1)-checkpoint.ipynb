```python
def dateSTR(x):
    if x < 2000:
        return 'Old'
    elif 2000 < x < 2010:
        return 'Modern'
    else:
        return 'New'


def dateENC(x):
    if x == 'Old':
        return 0
    elif x == 'Modern':
        return 1
    else:
        return 2

df['date']=df['date'].apply(lambda x: dateSTR(x))
```

```python
import gensim
import numpy as np

# Step 1: Load Word2Vec model (pre-trained)
word2vec_model_path = "C:/Users/Turbo/Desktop/GoogleNews-vectors-negative300.bin"
word2vec_model = gensim.models.KeyedVectors.load_word2vec_format(fname=word2vec_model_path, binary=True)


# Function to get text embedding for a movie description
def get_text_embedding(text, model):
    words = text.split()  # Tokenize the description
    word_vectors = [model[word] for word in words if word in model]
    if len(word_vectors) > 0:
        return np.mean(word_vectors, axis=0)
    else:
        return np.zeros(model.vector_size)  # Return zero vector if no words found
```

```python

# Step 2: Get embeddings for movie descriptions using Word2Vec
text_embeddings = np.array([get_text_embedding(desc[0], word2vec_model) for desc in df[['overview']].values])
```

```python
intxt_emb = np.array([get_text_embedding(overview, word2vec_model)])
input_combined_embeddings = np.hstack([inval_emb, intxt_emb])
input_combined_embeddings.shape
```

```python
import numpy as np
import faiss

# Assume 'movie_embeddings' is the embedding matrix you obtained from the encoder
# movie_embeddings should be of shape (n_movies, encoding_dim)

# Step 1: Create the FAISS index
d = combined_embeddings.shape[1]  # The dimensionality of the embeddings
index = faiss.IndexFlatL2(d)  # L2 distance index (Euclidean distance)

# Step 2: Add the movie embeddings to the FAISS index
index.add(combined_embeddings.astype(np.float32))  # FAISS requires float32 type

# Step 3: Search for similar movies
# Example: Find the 5 most similar movies to movie at index 0
k = 5  # Number of similar movies to retrieve
D, I = index.search(input_combined_embeddings.astype(np.float32), k)

# D contains the distances of the top k similar movies (lower is more similar)
# I contains the indices of the top k similar movies
  # Use the index to retrieve the movie title
print("Most similar movies to the new movie (Titles):")
for idx, dis in zip (I[0], D[0]):
    print(f"{titles[idx]} (Distance: {dis:.4f})")    # Use the index to retrieve the movie title
```
